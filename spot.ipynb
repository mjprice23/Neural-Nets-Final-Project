```python
def show_tracks(tracks):
    songs = []
    if tracks['items']:
        for i, item in enumerate(tracks['items']):
            if item['track']:
                track = item['track']
                if track['name']:
                    songs.append(track['name'])
    return songs
```

```python
def get_playlists(user):
    titles = []
    songlists = []
    playlists = sp.user_playlists(user)
    print(user)
    while playlists:
        for i, playlist in enumerate(playlists['items']):
            print("%4d %s %s" % (i + 1 + playlists['offset'], playlist['uri'],  playlist['name']))
            results = sp.playlist(playlist['id'],fields="tracks,next")
            tracks = results['tracks']
            songs = show_tracks(tracks)
            if len(songs) > 0:
                songlists.append(songs)
            else:
                songlists.append('None')
            titles.append(playlist['name'])
        if playlists['next']:
            playlists = sp.next(playlists)
        else:
            playlists = None
    return titles, songlists
```

```python
usernames = []
with open("usernames.csv", "r") as f:
	for line in f:
		data = line.strip()
		usernames.append(data)
```

```python
def get_data():
    titles = []
    tracks = []
    for user in usernames:
        playlists, songs = get_playlists(user)
        titles.extend(playlists)
        tracks.extend(songs)
    df = pd.DataFrame(list(zip(titles, tracks)), 
               columns =['Playlist Title', 'Track List'])
    print(df)
    df.to_csv("playlists.csv")
```
